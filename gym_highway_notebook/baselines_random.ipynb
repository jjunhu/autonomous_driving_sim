```python
import gymnasium as gym
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.vec_env import VecVideoRecorder
from stable_baselines3.common.env_util import make_vec_env
from HighwayEnv.scripts.utils import record_videos, show_videos

# Define a function to create and wrap the environment
def make_env():
    env = gym.make('highway-v0', render_mode='rgb_array')
    return env

# Create vectorized environment
env = make_env()
env = record_videos(env, video_folder="baseline_highway_v0")

obs = env.reset()
for _ in range(10):  # Adjust the range or conditions for longer or different behaviors
    # Select a random action from the environment's action space
    action = env.action_space.sample()
    obs, rewards, dones, truncated, info = env.step(action)
    if dones:
        obs = env.reset()

env.close()
show_videos()
```
